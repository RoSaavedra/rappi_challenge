```python
def get_data():
    df = cp_fraud = pd.read_csv("https://bitbucket.org/fernandopaulin/rpmx_ds_challenge/raw/9c6f6f0dd29e7d8a1a720590f2190ffaf47a4790/ds_challenge_apr2021.csv", sep=',')
    return df
```

```python
def tad(cp_fraud):
    time = pd.to_datetime(pd.to_datetime(cp_fraud.fecha, errors='coerce'))
    cp_fraud['dia_semana'] = time.dt.dayofweek
    #dia del mes
    cp_fraud['dia_mes'] = time.dt.day
    #month
    cp_fraud['mes'] = time.dt.month
    #year
    cp_fraud['anio'] = time.dt.year
    #datos del dispositivo
    cp_fraud['dispositivo'] = cp_fraud.dispositivo.map(lambda x:ast.literal_eval(x))
    cp_fraud = pd.concat([cp_fraud, json_normalize(cp_fraud.dispositivo)], axis =1)
    #creamos factor dummie para datos faltantes
    cp_fraud.establecimiento.fillna('sin_info', inplace = True)
    #creamos factor dummie para datos faltantes
    cp_fraud.ciudad.fillna('sin_info', inplace = True)
    data = pd.get_dummies(data = cp_fraud, 
                          columns = ['genero', 
                                     'establecimiento',
                                     'ciudad', 
                                     'tipo_tc', 
                                     'status_txn', 
                                     'is_prime',
                                     'os'],
                          prefix_sep='_', 
                          drop_first= False)

    data.reset_index(inplace = True)
    data.drop(columns = 'index', inplace = True)
        
    return data
```

```python
def modelado(df, modelo, scaler):
    
    columns=['monto', 'hora', 'linea_tc', 'interes_tc', 'dcto', 'cashback', 
       'dia_semana', 'dia_mes', 'device_score', 
        'genero_F', 'genero_M', 'establecimiento_Farmacia',
       'establecimiento_MPago', 'establecimiento_Restaurante',
       'establecimiento_Super', 'establecimiento_sin_info', 'ciudad_Merida',
       'ciudad_Monterrey', 'ciudad_Toluca', 'ciudad_sin_info',
       'tipo_tc_Virtual', 'status_txn_En proceso', 'status_txn_Rechazada',
       'is_prime_True', 'os_.', 'os_ANDROID', 'os_WEB']
    
    
    json={}
    json['timestamp'] = '12/06/2021T10:00:00'
    json['aplicante'] = 'mrsp'  
    json['predict'] = modelo.predict(scaler.fit_transform(df[columns]))
        
    return json
```
